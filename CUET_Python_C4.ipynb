```python
#sum 1 to 100 using for loop
sum = 0
for i in range(1, 101):
  sum += i
print("The sum of numbers from 1 to 100 is:", sum)
```

```python
#nested for loop
for i in range(0,1):
    for j in range (0,2):
        for k in range (0,3):
            for m in range (0,4):
                print(f'i={i},j={j},k={k},m={m}')
```

**CONTROL** **STATEMENT**
---

```python
#break statement
var = 5
while var == 5:
  num = int(input("Enter a number: "))
  print("You entered: ", num)
  break
  print("End of while loop")
```

```python
#break  statement
for letter in 'Kabyo':
    if letter == 'b':
        break
    print('Current letter;',letter)
print('dhur jata')
```
